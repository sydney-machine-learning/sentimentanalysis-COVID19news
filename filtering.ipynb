```python
import pandas as pd

def filter_and_combine_australia_sections(df, start_date, end_date, section_names, output_file_path):
    """
    根据指定日期范围和多个部分名称筛选数据，并保存到指定路径。

    Args:
    - df: DataFrame
    - start_date: 起始日期，格式为 'YYYY-MM-DD'
    - end_date: 结束日期，格式为 'YYYY-MM-DD'
    - section_names: 部分名称的列表
    - output_file_path: 输出文件路径
    """
    # 转换日期列为日期时间格式
    df['webPublicationDate'] = pd.to_datetime(df['webPublicationDate'])

    # 转换起始和结束日期为日期时间格式，并进行时区本地化
    start_date_utc = pd.to_datetime(start_date).tz_localize('UTC')
    end_date_utc = pd.to_datetime(end_date).tz_localize('UTC')

    # 根据section_names筛选数据
    filtered_df = df[(df['webPublicationDate'] >= start_date_utc) & 
                     (df['webPublicationDate'] <= end_date_utc) & 
                     (df['sectionName'].isin(section_names))]

    # 删除缺失内容的行
    filtered_df = filtered_df.dropna(subset=['bodyContent'])

    # 保存数据
    filtered_df.to_csv(output_file_path, index=False)

    print("Filtered data has been saved to:", output_file_path)

# 读取CSV文件到DataFrame
df = pd.read_csv('/Users/neo/Documents/ngrams/guardian_articles.csv')  # 请替换为实际的文件路径

# 设置日期范围和部分名称列表
start_date = '2019-12-31'
end_date = '2022-03-25'
australia_sections = [
    'Australia news', 
    'Guardian Live Australia', 
    'Dairy Australia: Listen to your gut', 
    'Dairy Australia: Sustainable futures'
]
section_names = ['Business', 'Healthcare Professionals Network','Money','Politics','Guardian Sustainable Business','Social Care Network','Personal Investments','Business to business']  # 根据需要调整部分名称
health_economy_output_file_path = "/Users/neo/Documents/ngrams/health_economy_articles_2018_to_2022.csv" # 请替换为实际路径



# 定义输出文件路径
au_news_output_file_path = "/Users/neo/Documents/ngrams/australia_news_articles_2018_to_2022.csv"

# 执行筛选和合并操作
filter_and_combine_australia_sections(df, start_date, end_date, australia_sections, au_news_output_file_path)


# 执行筛选和合并操作
filter_and_combine_australia_sections(df, start_date, end_date, section_names, health_economy_output_file_path)

```

```python
import pandas as pd
from datetime import datetime

df = pd.read_csv('/Users/neo/Documents/ngrams/guardian_articles.csv')
df['webPublicationDate'] = df['webPublicationDate'].apply(
    lambda x: datetime.strptime(x, "%Y-%m-%dT%H:%M:%SZ").strftime("%Y-%m-%d")
)

df.head()
columns_to_drop = ['article_id', 'webTitle', 'webUrl', 'id']
modified_df = df.drop(columns=columns_to_drop, axis=1)
```
